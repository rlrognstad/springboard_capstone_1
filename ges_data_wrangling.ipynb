Read in GES data from csv (2015) or txt (all other years) files 

```python
accident_15 = pd.DataFrame.from_csv('nhtsa-ges-data/GES_2015/ACCIDENT.CSV', index_col=None)
accident_15['YEAR'] = 2015 #add column to specify year, needed when compling multiyear dataset
#accident_15.head()
person_15 = pd.DataFrame.from_csv('nhtsa-ges-data/GES_2015/PERSON.CSV', index_col=None)
person_15['YEAR'] = 2015 #add column to specify year, needed when compling multiyear dataset
#person_15.head()

vehicle_15 = pd.DataFrame.from_csv('nhtsa-ges-data/GES_2015/VEHICLE.CSV', index_col=None, encoding = 'latin-1')
vehicle_15['YEAR'] = 2015 #add column to specify year, needed when compling multiyear dataset
vehicle_15.head()
```

Read in .txt files, which are tab-delimited and convert to data frames

```python
accident_14 = pd.read_csv('nhtsa-ges-data/GES_2014/ACCIDENT.TXT',sep='\t',header=(0))
accident_14['YEAR'] = 2014 #add column to specify year, needed when compling multiyear dataset
#accident_14.head()
person_14 = pd.read_csv('nhtsa-ges-data/GES_2014/PERSON.TXT',sep='\t',header=(0))
person_14['YEAR'] = 2014 #add column to specify year, needed when compling multiyear dataset
#person_14.head()

vehicle_14 = pd.read_csv('nhtsa-ges-data/GES_2014/VEHICLE.TXT',sep='\t',header=(0))
vehicle_14['YEAR'] = 2014 #add column to specify year, needed when compling multiyear dataset
#vehicle_14.head()
```

```python
accident_13 = pd.read_csv('nhtsa-ges-data/GES_2013/ACCIDENT.TXT',sep='\t',header=(0))
accident_13['YEAR'] = 2013 #add column to specify year, needed when compling multiyear dataset
#accident_13.head()
person_13 = pd.read_csv('nhtsa-ges-data/GES_2013/PERSON.TXT',sep='\t',header=(0))
person_13['YEAR'] = 2013 #add column to specify year, needed when compling multiyear dataset
#person_13.head()

vehicle_13 = pd.read_csv('nhtsa-ges-data/GES_2013/VEHICLE.TXT',sep='\t',header=(0))
vehicle_13['YEAR'] = 2013 #add column to specify year, needed when compling multiyear dataset
vehicle_13.head()
```

```python
accident_12 = pd.read_csv('nhtsa-ges-data/GES_2012/ACCIDENT.TXT',sep='\t',header=(0))
accident_12['YEAR'] = 2012 #add column to specify year, needed when compling multiyear dataset
#accident_12.head()
person_12 = pd.read_csv('nhtsa-ges-data/GES_2012/PERSON.TXT',sep='\t',header=(0))
person_12['YEAR'] = 2012 #add column to specify year, needed when compling multiyear dataset
#person_12.head()

vehicle_12 = pd.read_csv('nhtsa-ges-data/GES_2012/VEHICLE.TXT',sep='\t',header=(0))
vehicle_12['YEAR'] = 2012 #add column to specify year, needed when compling multiyear dataset
#vehicle_12.head()
```

```python
accident_11 = pd.read_csv('nhtsa-ges-data/GES_2011/ACCIDENT.TXT',sep='\t',header=(0))
accident_11['YEAR'] = 2011 #add column to specify year, needed when compling multiyear dataset
#accident_11.head()
person_11 = pd.read_csv('nhtsa-ges-data/GES_2011/PERSON.TXT',sep='\t',header=(0))
person_11['YEAR'] = 2011 #add column to specify year, needed when compling multiyear dataset
#person_11.head()

vehicle_11 = pd.read_csv('nhtsa-ges-data/GES_2011/VEHICLE.TXT',sep='\t',header=(0))
vehicle_11['YEAR'] = 2011 #add column to specify year, needed when compling multiyear dataset
#vehicle_11.head()
```

```python
#make combined dataframes
a_frames = [accident_15, accident_14, accident_13, accident_12, accident_11]
accident_all = pd.concat(a_frames)

accident_all.head()
```

```python
p_frames = [person_15, person_14, person_13, person_12, person_11]
person_all = pd.concat(p_frames)

person_all.head()
```

```python
v_frames = [vehicle_15, vehicle_14, vehicle_13, vehicle_12, vehicle_11]
vehicle_all = pd.concat(v_frames)

vehicle_all.head()
```

## Import census data, 2011-2015

```python
census = pd.DataFrame.from_csv('census_data/cc-est2015-subset.csv', index_col=None, encoding = 'utf-8')
census = census[census['YEAR'] >= 4] #filter out desired census years, 2011:2015
#convert coded census years to actual years, 4=2011; 5=2012; 6=2013; 7=2014; 8=2015
census['YEAR'] = census['YEAR'] + 2007
census.head()
```
